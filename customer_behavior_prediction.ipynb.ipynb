```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data=pd.read_csv("Customer Purchasing Behaviors.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.drop(columns=['purchase_frequency', 'user_id', 'age'], errors='ignore')
y = data['purchase_frequency']
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['region']=le.fit_transform(data['region'])
data
```

```python
x=data.iloc[:,:-2]
print(x)
y=data['purchase_frequency']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
x=data.iloc[:,:-1]
y=data['purchase_frequency']
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector
lr=LinearRegression()
fs=SequentialFeatureSelector(lr,k_features=3,forward=True,scoring='r2', cv=5)
fs.fit(x_train,y_train)
```

```python
selected_features = list(fs.k_feature_names_)
x_train_fs = x_train[selected_features]
x_test_fs = x_test[selected_features]
```

```python
lr.fit(x_train_fs, y_train)
```

```python
y_pred = lr.predict(x_test_fs)
```

```python
print("Selected Features:", selected_features)
print("R² Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
```
